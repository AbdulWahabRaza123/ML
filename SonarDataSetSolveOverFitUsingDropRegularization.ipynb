```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#I am using the header=None because we don't have column name in the dataset so to set them in number wise I use header=None
df=pd.read_csv("/content/sonar.all-data.csv",header=None)
df.sample(5)
```

```python
df.shape
```

```python
#to check the NA value by making sum to every row
df.isna().sum()
```

```python
#column 60 is y column which is giving us actual results and we use the value_counts to get the results of every row
df[60].value_counts()
```

```python
#we got two labels only so this is binary classification problem
X=df.drop(60,axis='columns') # by using axis=columns we can drop the column 60 instead of row
y=df[60]
```

```python
X
```

```python
y
```

```python
#Now we will use one hot encoding for y
y=pd.get_dummies(y,drop_first=True) #we using drop_first to remove extra columns during dummy creation
y.sample(4)
```

```python
y.value_counts()
```

```python
#Now using train and test split library
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
#Now creating model
tf.random.set_seed(42)
model=tf.keras.Sequential(
    [
        tf.keras.layers.Dense(60,activation='relu'),
     tf.keras.layers.Dense(30,activation='relu'),
     tf.keras.layers.Dense(15,activation='relu'),
     tf.keras.layers.Dense(1,activation='sigmoid')
    ]
)
#model compilation
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
#fitting model
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)#our model is overfit
```

```python
#Now predict some of the values
y_pred=model.predict(X_test).reshape(-1)#we are using reshape(-1) to convert them to 1d array
y_pred=np.round(y_pred)
y_pred[:10]
```

```python
y_test[:10]
```

```python
#Now we will create the classification report
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```

###To reduce the Overftting in the model we can use dropout layer

```python
#Now creating model
tf.random.set_seed(42)
model1=tf.keras.Sequential(
    [
        tf.keras.layers.Dense(60,input_dim=60,activation='relu'),
     tf.keras.layers.Dropout(0.2),#20% layers will be drop out here
     tf.keras.layers.Dense(30,activation='relu'),
     tf.keras.layers.Dropout(0.2),#20% layers will be drop out here
     tf.keras.layers.Dense(15,activation='relu'),
     tf.keras.layers.Dropout(0.2),#20% layers will be drop out here
     tf.keras.layers.Dense(1,activation='sigmoid')
    ]
)
#model compilation
model1.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
#fitting model
model1.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model1.evaluate(X_test,y_test)
```
